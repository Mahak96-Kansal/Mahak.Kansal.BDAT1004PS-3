Occupations
Step 1. Import the necessary libraries
Step 2. Import the dataset from this address. 
Step 3. Assign it to a variable called users
Step 4. Discover what is the mean age per occupation
Step 5. Discover the Male ratio per occupation and sort it from the most to the least 
Step 6. For each occupation, calculate the minimum and maximum ages
Step 7. For each combination of occupation and sex, calculate the mean age 
Step 8. For each occupation present the percentage of women and men

```python
# Step 1: Import the necessary libraries
import pandas as pd
```

```python
# Step 2: Import the dataset from the provided address
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, delimiter='|')
```

```python
# Step 3: Assign it to a variable called users (already assigned as 'users')
users.head()
```

```python
# Step 4: Discover what is the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean age per occupation:")
print(mean_age_per_occupation)
```

```python
# Step 5: Discover the Male ratio per occupation and sort it from the most to the least
male_ratio_per_occupation = users[users['gender'] == 'M'].groupby('occupation').size() / users.groupby('occupation').size()
male_ratio_per_occupation_sorted = male_ratio_per_occupation.sort_values(ascending=False)
print("\n Male ratio per occupation :")
print(male_ratio_per_occupation_sorted)
```

```python
# Step 6: For each occupation, calculate the minimum and maximum ages
min_max_age_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nMinimum and Maximum age per occupation:")
min_max_age_per_occupation
```

```python
# Step 7: For each combination of occupation and sex, calculate the mean age
mean_age_per_occupation_sex = users.groupby(['occupation', 'gender'])['age'].mean().unstack()
print("\nMean age per occupation and gender:")
mean_age_per_occupation_sex
```

```python
# Step 8: For each occupation, present the percentage of women and men
gender_percentage_per_occupation = users.groupby('occupation')['gender'].value_counts(normalize=True).unstack().fillna(0) * 100
print("\nPercentage of women and men per occupation:")
gender_percentage_per_occupation
```

```python
# Step 1: Import the necessary libraries
import pandas as pd
import numpy as np
```

```python
# Step 2: Import the dataset
# Load dataset
df_wind = pd.read_csv("C:/Users/mahak kansal/Downloads/wind.txt", delim_whitespace=True)
data = df_wind
data.head()
```

```python
# Step 3: Assign it to a variable called `data` and replace the first 3 columns with a proper datetime index
data['Date'] = pd.to_datetime(data[['Yr', 'Mo', 'Dy']].astype(str).agg('-'.join, axis=1), format='%y-%m-%d')
data = data.set_index('Date')
data = data.drop(columns=['Yr', 'Mo', 'Dy'])
data.head()
```

```python
# Step 4: Year 2061? Create a function to fix it and apply it
def fix_year(date):
    if date.year > 2000:
        return date.replace(year=date.year - 100)
    return date

data.index = data.index.map(fix_year)
```

```python
# Step 5: Set the right dates as the index, ensuring datetime64[ns] type
data.index = pd.to_datetime(data.index)
type(data.index)
```

```python
# Step 6: Compute how many values are missing for each location over the entire record
missing_values = data.isnull().sum()
print("Step 6: Missing values for each location:\n", missing_values)
```

```python
# Step 7: Compute how many non-missing values there are in total
non_missing_values = data.notnull().sum().sum()
print("Total non-missing values:", non_missing_values)
```

```python
# Step 8: Calculate the mean windspeeds over all the locations and times
overall_mean = data.mean().mean()
print("Mean windspeed over all locations and times:", overall_mean)
```

```python
# Step 9: Create a DataFrame `loc_stats` with min, max, mean, and std for each location
loc_stats = data.aggregate(['min', 'max', 'mean', 'std']).transpose()
print("Statistics for each location:\n", loc_stats)
```

```python
# Step 10: Create a DataFrame `day_stats` with min, max, mean, and std across all locations for each day
day_stats = data.aggregate(['min', 'max', 'mean', 'std'], axis=1)
print("Statistics for each day:\n", day_stats)
```

```python
# Step 11: Find the average windspeed in January for each location
january_data = data[data.index.month == 1]
january_avg = january_data.mean()
print("Average windspeed in January for each location:\n", january_avg)
```

```python
# Step 12: Downsample the record to a yearly frequency for each location
yearly_data = data.resample('Y').mean()
print("Step 12: Yearly frequency data:\n", yearly_data)
```

```python
# Step 13: Downsample the record to a monthly frequency for each location
monthly_data = data.resample('M').mean()
print("Monthly frequency data:\n", monthly_data)
```

```python
# Step 14: Downsample the record to a weekly frequency for each location
weekly_data = data.resample('W').mean()
print("Weekly frequency data:\n", weekly_data)
```

```python
# Step 15: Calculate weekly min, max, mean, and std for each week (first week starts Jan 2, 1961) for the first 52 weeks
weekly_stats = data['1961-01-02':'1962-01-01'].resample('W').agg(['min', 'max', 'mean', 'std']).head(52)
print("Weekly statistics for the first 52 weeks:\n", weekly_stats)
```

## Question 6
Create a line plot showing the number of marriages and divorces per capita in the
U.S. between 1867 and 2014. Label both lines and show the legend. Don't forget to label your axes!

```python
# Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset 
df = pd.read_csv("C:/Users/mahak kansal/Downloads/us-marriages-divorces-1867-2014.csv")

# Check the first 5 rows
df.head()

# Example data
years = df['Year'] 
marriages = df['Marriages_per_1000']
divorces =  df['Divorces_per_1000']

# Create the line plot
plt.figure(figsize=(10, 6))

# Plot marriages and divorces per capita
plt.plot(years, marriages, label="Marriages per capita")
plt.plot(years, divorces, label="Divorces per capita")

# Label both lines and show the legend
plt.xlabel('Year')
plt.ylabel('Rate per Capita')
plt.title('Marriages and Divorces per Capita in the U.S. (1867-2014)')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```

## Question 7
Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python
# Filter the dataset for the years 1900, 1950, and 2000
selected_years = df[df['Year'].isin([1900, 1950, 2000])]

# Set up the bar chart data
labels = selected_years['Year']
marriages = selected_years['Marriages_per_1000']
divorces = selected_years['Divorces_per_1000']

# Create the bar chart
width = 0.35  # Bar width
x = range(len(labels))

fig, ax = plt.subplots(figsize=(10, 6))

# Plot marriages and divorces as vertical bars
bar1 = ax.bar(x, marriages, width, label='Marriages per capita', color='pink')
bar2 = ax.bar([p + width for p in x], divorces, width, label='Divorces per capita', color='skyblue')

# Label axes and add a legend
ax.set_xlabel('Year')
ax.set_ylabel('Rate per 1000 People')
ax.set_title('Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
ax.set_xticks([p + width / 2 for p in x])  # Center the x-ticks between the two bars
ax.set_xticklabels(labels)
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
